```python
import os
from gprMax.gprMax import api
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to read GPRMax HDF5 output file
def read_gprmax_h5(file_path):
    with h5py.File(file_path, 'r') as f:
        # Read the data from the file
        data = f['rxs/rx1/Ez'][:]
    return data

# Function to convert .out file to .png
def convert_bscan_to_png(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(input_folder):
        if filename.endswith(".out"):
            file_path = os.path.join(input_folder, filename)
            bscan_data = read_gprmax_h5(file_path)

            plt.figure(figsize=(10, 6))
            plt.imshow(bscan_data.T, aspect='auto', cmap='gray', extent=[0, bscan_data.shape[0], bscan_data.shape[1], 0])
            plt.colorbar(label='Amplitude')
            plt.xlabel('Trace number')
            plt.ylabel('Time step')
            plt.title('B-scan')

            output_file = os.path.join(output_folder, filename.replace('.out', '.png'))
            plt.savefig(output_file)  # Save the plot as a PNG file
            plt.close()

# Define input and output folders
input_folder = '/content/Interns/COD_Dataset/No_Cyl_output'
output_folder = '/content/Interns/COD_Dataset/No_Cyl_images'

# Convert all .out files in the input folder to .png files in the output folder
convert_bscan_to_png(input_folder, output_folder)
```
